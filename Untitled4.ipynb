```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_data = ImageDataGenerator(rescale=1/255,
                                shear_range=0.2,
                                zoom_range=0.2,
                                horizontal_flip=True)
val_data = ImageDataGenerator(rescale=1/255)
test_data = ImageDataGenerator(rescale=1/255)
```

```python
train_gen = train_data.flow_from_directory('./images/train',
                                           color_mode='rgb',
                                           class_mode='categorical',
                                           batch_size=32,
                                           shuffle=True)
val_gen = val_data.flow_from_directory('./images/val',
                                       shuffle=False)
test_gen = val_data.flow_from_directory('./images/test',
                                        shuffle=False)
```

```python
img, labels = train_gen.next()
```

```python
plt.figure(figsize=(15,20))
for i in range(len(img)):
    plt.subplot(8,4,i+1)
    plt.title(labels[i])
    plt.imshow(img[i])
    plt.axis('off')
plt.tight_layout()
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
base_model = VGG16(input_shape=(256,256,3),
                   include_top=False,)
```

```python
for layer in base_model.layers:
    layer.trainable=False
```

```python
x = tf.keras.layers.Flatten()(base_model.output)
x = tf.keras.layers.Dense(512, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(4, activation='softmax')(x)

model = tf.keras.models.Model(base_model.input, x)

model.compile(optimizer=tf.keras.optimizers.RMSprop(lr=1e-4),
              loss='categorical_crossentropy',
              metrics=['acc'])

model.summary()
```

```python
hist = model.fit(train_gen,
                 validation_data=val_gen,
                 steps_per_epoch=100,
                 epochs=25)
```

```python
model.evaluate(test_gen)
```

```python
y_true = test_gen.classes
y_pred = model.predict(test_gen)
```

```python
def build_model(summary=False):
    x = tf.keras.layers.Flatten()(base_model.output)
    x = tf.keras.layers.Dense(512, activation='relu')(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    x = tf.keras.layers.Dense(4, activation='softmax')(x)

    model = tf.keras.models.Model(base_model.input, x)

    model.compile(optimizer=tf.keras.optimizers.RMSprop(lr=1e-4),
                  loss='categorical_crossentropy',
                  metrics=['acc'])
    if summary:
        model.summary()
    
    return model
```

```python
def build_generators(i):  
    train_gen = train_data.flow_from_dataframe(train_df,
                                               directory=None,
                                               x_col=i,
                                               y_col='label')
    val_gen = val_data.flow_from_dataframe(val_df,
                                             directory=None,
                                             x_col=i,
                                             y_col='label')
    test_gen = test_gen.flow_from_dataframe(test_df,
                                              directory=None,
                                              x_col=i,
                                              y_col='label',
                                              shuffle=False)
    
    return train_gen, val_gen, test_gen

def fit_model(model, epochs=5):
    hist = model.fit(train_gen,
                     validation_data=val_gen,
                     epochs=epochs)
    
    return hist

def make_col(model, i):
    preds = model.predict(train_gen)
    pred_col = [list(x) for x in preds]
    pred_col = pd.Series(pred_col, name=str(i))
    
    return pred_col

def add_col(col_name, df, data):
    df[str(col_name)] = data
    
```

```python
img_count = 10
epochs = 3
pred_df = pd.DataFrame()
pred_array = np.empty((len(test_df), img_count, 4))
for i in range(img_count):
    print('image: {}'.format(i))
    train_gen, val_gen, test_gen = build_generators(i)
    model = build_model()
    hist = fit_model(model, epochs=epochs)
    model.evaluate(test_gen)
    pred_col = make_col(model, i)
    add_col(i, pred_df, pred_col)
    pred_array[:,i,:] = model.predict(test_gen)
    print()
    print('-----' * 5)
    print()
```
