```python
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_data = ImageDataGenerator(rescale=1/255,
                                horizontal_flip=True)
val_data = ImageDataGenerator(rescale=1/255)
test_data = ImageDataGenerator(rescale=1/255)
```

```python
train_generator = train_data.flow_from_directory('images/train',
                                                 color_mode='rgb',
                                                 target_size=(224,398),
                                                 batch_size=32,
                                                 class_mode='categorical')
val_generator = val_data.flow_from_directory('images/val',
                                             color_mode='rgb',
                                             target_size=(224,398),
                                             batch_size=32,
                                             class_mode='categorical')
test_generator = test_data.flow_from_directory('images/test',
                                               color_mode='rgb',
                                               target_size=(224,398),
                                               batch_size=32,
                                               class_mode='categorical')
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def init_callbacks():
    early_stopping = EarlyStopping(monitor='val_loss', patience=5)
    model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss',
                                      verbose=1, save_best_only=True,
                                      save_weights_only=True)
    
    callbacks = [early_stopping, model_checkpoint]
    return callbacks
```

```python
model_1 = Sequential()
model_1.add(Dense(128, activation='relu', input_shape=(224,398,3)))
model_1.add(Flatten())
model_1.add(Dense(4, activation='softmax'))

model_1.compile(loss='categorical_crossentropy',
                optimizer='adam',
                metrics=['acc'])
model_1.summary()
```

```python
history_1 = model_1.fit(train_generator,
                        epochs=25,
                        validation_data=val_generator,
                        callbacks=init_callbacks())
```

```python
model_1.evaluate(test_generator)
```

```python
predictions = model_1.predict(test_generator)
predictions[:5]
```
