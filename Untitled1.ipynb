```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.metrics import confusion_matrix, classification_report

import numpy as np
import pandas as pd
```

```python
train_data = ImageDataGenerator(rescale=1/255,
                                horizontal_flip=True)
val_data = ImageDataGenerator(rescale=1/255)
test_data = ImageDataGenerator(rescale=1/255)
```

```python
train_generator = train_data.flow_from_directory('images/train',
                                                 color_mode='rgb',
                                                 target_size=(224,398),
                                                 batch_size=32,
                                                 shuffle=True,
                                                 class_mode='categorical')
val_generator = val_data.flow_from_directory('images/val',
                                             color_mode='rgb',
                                             target_size=(224,398),
                                             batch_size=32,
                                             shuffle=False,
                                             class_mode='categorical')
test_generator = test_data.flow_from_directory('images/test',
                                               color_mode='rgb',
                                               target_size=(224,398),
                                               batch_size=32,
                                               shuffle=False,
                                               class_mode='categorical')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img, labels = train_generator.next()

plt.figure(figsize=(15,20))
for i in range(len(img)):
    plt.subplot(8,4,i+1)
    plt.title(labels[i])
    plt.imshow(img[i])
    plt.axis('off')
    plt.colorbar()
plt.tight_layout()
```

```python
def init_callbacks():
    early_stopping = EarlyStopping(monitor='val_loss', patience=5)
    model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss',
                                     verbose=1, save_best_only=True,
                                     save_weights_only=True)

    callbacks = [early_stopping, model_checkpoint]
    return callbacks
```

```python
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(3,3), padding='same',
                 activation='relu', input_shape=(224,398,3)))

model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3,3), padding='same', activation='relu'))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(128, (3,3), padding='same', activation='relu'))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(Activation('relu'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(512, (5,5), padding='same', activation='relu'))
model.add(Conv2D(512, (5,5), activation='relu'))
model.add(MaxPooling2D((4,4)))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(4, activation='softmax'))

optimizer = tf.keras.optimizers.RMSprop(learning_rate=10e-4)

model.compile(loss='categorical_crossentropy',
              optimizer=optimizer,
              metrics=['acc'])
model.summary()
```

```python
history = model.fit(train_generator,
                    epochs=25,
                    validation_data=val_generator,
                    callbacks=init_callbacks())
```

```python
def plot_model(hist):
    '''
    input:model history
    output: plots of accuracy and loss for train and val sets
    '''
    history = hist.history
    x = range(1,len(hist.history['loss'])+1)
    fig, ax = plt.subplots(1,2,figsize=(15,5))
    ax[0].plot(x, history['acc'], label='acc')
    ax[0].plot(x, history['val_acc'], label='val_acc')
    ax[0].legend()
    ax[0].set_xticks(range(1,(len(history['acc']))+1),5)
    ax[0].set_title('Accuracy')
    ax[1].plot(x, history['loss'], label='loss')
    ax[1].plot(x, history['val_loss'], label='val_loss')
    ax[1].legend()
    ax[1].set_xticks(range(1,(len(history['acc']))+1),5)
    ax[1].set_title('Loss')
    plt.show()
```

```python
plot_model(history)
```
